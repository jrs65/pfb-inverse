## PFB

As described by Kevin, the PFB essentially does the follow:

- Take an 8192 length chunk
- Apply a window function
- FFT the chunk
- Take every 4th frequency
- Shift by 2048 samples and repeat

There's also a truncation of the output frequencies done by throwing away the Nyquist frequency, which reduces the number of output frequencies down to 1024.

Mathematically, let's denote the full length chunk which is $N$ samples long as $X_a(p)$, and the window function is $w(p)$. The above process can be written as
\begin{equation}
X_a(f) = \sum_p x_a(p) w(p) e^{-2 \pi i \alpha f p / N}
\end{equation}
where $M$ is the decimation factor. To move to the next spectrum we shift by an amount $n = N / \alpha$. The number of frequencies we get out of this process is $\sim n / 2$. This is pretty straightforward to implement. First we define the window function we will use.

```python
def sinc_window(ntap, lblock):
    """Sinc window function.
    
    Parameters
    ----------
    ntaps : integer
        Number of taps.
    lblock: integer
        Length of block.
        
    Returns
    -------
    window : np.ndarray[ntaps * lblock]
    """
    coeff_length = np.pi * ntap
    coeff_num_samples = ntap * lblock
    
    # Sampling locations of sinc function
    X = np.arange(-coeff_length / 2.0, coeff_length / 2.0,
                  coeff_length / coeff_num_samples)
    
    # np.sinc function is sin(pi*x)/pi*x, not sin(x)/x, so use X/pi
    return np.sinc(X / np.pi)


def sinc_hanning(ntap, lblock):
    """Hanning-sinc window function.
    
    Parameters
    ----------
    ntaps : integer
        Number of taps.
    lblock: integer
        Length of block.
        
    Returns
    -------
    window : np.ndarray[ntaps * lblock]
    """
    
    return sinc_window(ntap, lblock) * np.hanning(ntap * lblock)
```

The we define the routine to actually perform the transform.

```python
def pfb(timestream, nfreq, ntap=4, window=sinc_hanning):
    """Perform the CHIME PFB on a timestream.
    
    Parameters
    ----------
    timestream : np.ndarray
        Timestream to process
    nfreq : int
        Number of frequencies we want out (probably should be odd
        number because of Nyquist)
    ntaps : int
        Number of taps.

    Returns
    -------
    pfb : np.ndarray[:, nfreq]
        Array of PFB frequencies.
    """
    
    # Number of samples in a sub block
    lblock = 2 * (nfreq - 1)
    
    # Number of blocks
    nblock = timestream.size / lblock - (ntap - 1)
    
    # Initialise array for spectrum
    spec = np.zeros((nblock, nfreq), dtype=np.complex128)
    
    # Window function
    w = window(ntap, lblock)
    
    # Iterate over blocks and perform the PFB
    for bi in range(nblock):
        # Cut out the correct timestream section
        ts_sec = timestream[(bi*lblock):((bi+ntap)*lblock)].copy()
        
        # Perform a real FFT (with applied window function)
        ft = np.fft.rfft(ts_sec * w)
        
        # Choose every n-th frequency
        spec[bi] = ft[::ntap]
        
    return spec
```

## Inverse PFB

To try and construct a suitable inverse for the PFB let's start with it's definition as write above:
\begin{equation}
X_a(f) = \sum_p x_a(p) w(p) e^{-2 \pi i \alpha f p / N}
\end{equation}

First step is to try perform the FFT on each spectrum to create a pseudo-timestream $x_a(q)$
\begin{align}
\tilde{x}_a(q) = \frac{1}{n} \sum_f X_a(f) e^{2 \pi i f q / n}
\end{align}
where as a reminder $n = N / M$. This transformation is completely loss-less and so we are free to work in this space. Going through the manipulation
\begin{align}
\tilde{x}_a(q) & = \frac{1}{n} \sum_f X_a(f) e^{2 \pi i f q / n} \\
& = \frac{1}{n} \sum_{f, p} w(p) x_a(p) e^{2 \pi i f (q - p) / n} \\
& = \sum_{\alpha=0}^{M-1} w(p) \delta_{q,p + \alpha n} x_a(p)
\end{align}

That is that each point in the pseudo-timestream is the summation over $M$ points in the real timestream, weighted by the value of the window function at that location.

Combining these for all the offset blocks in the timestream generates a sparse matrix mapping one timestream into the other. This matrix has only $M$ non-zero diagonals. This matrix is also rectangular, as there are more entries in the real timstream than the pseudo-timestream, and thus we can not exactly invert to get to the real timestream.

```python
def pfb_timestream_fullmatrix(ntime, nfreq, ntap=4, window=sinc_hanning):
    
    # Number of samples in a sub-block
    lblock = 2*(nfreq - 1)
    
    # Number of blocks in timestream
    nblocks = ntime / lblock
    
    # Number of blocks in PFB
    npfb = nblocks - ntap + 1
    
    # Initialise matrix
    mat = np.zeros((npfb, lblock, nblocks, lblock))
    
    # Window function
    w = window(ntap, lblock)
    
    # Iterate over PFB blocks setting the elements
    for bi in range(npfb):
        for si in range(lblock):
            for ai in range(ntap):
                mat[bi, si, bi+ai, si] = w[si + ai * lblock]
    
    return mat
```

```python
pfm = pfb_timestream_fullmatrix(2048, 17)
```

```python
pfmr = pfm.reshape(-1, 2048)
```

```python
imshow(np.log10(np.abs(pfmr[:400, :400])), interpolation='nearest')
colorbar()
```

```python
# Generate inversion matrix

ipfmr = la.pinv2(pfmr)
```

```python
# Try and reconstruct timestream

pseudo_ts = np.fft.irfft(spec_pfb, axis=1).flatten()
recovered_ts = np.dot(ipfmr, pseudo_ts)
```

Clearly we can reconstruct the timestream pretty well (except at the ends). The accuracy is mostly dependent on the number of samples the pseudo-timestream is missing (and so becomes much worse if we go to more frequencies for the same number of timestream points). I haven't included it here, but you can show that the residuals are exactly equal to the original timestream projected onto the PFB matrix nullspace, so I don't think you can ever do better.
 
The above recovery is computationally expensive because we need to generate the pseudo-inverse for the whole timestream which takes $O(N^3)$ time. However, as the PFB transformation matrix is band diagonal, we should be able to perform the inverse in $O(N)$ time.

## Fast PFB Inversion

The PFB matrix maps between blocks of length $n$ from the timestream into the pseudo-timestream. Let's make this explicit by writing the timestream index as
\begin{equation}
q = b n + d
\end{equation}
where $d$ ranges from from $1 \ldots n-1$. Similarly the pseudo timestream index is
\begin{equation}
p = a n + c \; .
\end{equation}
With this in mind we can write the PFB mapping matrix in terms of a compound indices $(ac)$ instead of $p$ and $(bd)$ instead of $q$, giving
\begin{align}
P_{(ac)(bd)} & = \sum_{\alpha = 0}^{M-1} \delta(a n + c + \alpha n, b n + d) w_\alpha(d) \\
& = \sum_{\alpha=0}^{M-1} \delta(a + \alpha, b) \delta(c, d) w_\alpha(d)
\end{align}
where $w_\alpha(d)$ corresponds to each length $n$ block of the length $N$ window function.

There is no particular reason we must consider the timestream in the natural order, and if in particular we imagine re-indexing it such that the block index was fastest moving in the matrix above (effectively transposing length $n$ blocks) we can see that $\delta(c,d)$ term makes this new matrix manifestly block diagonal. This allows us to break the full inverse problem into $n$ individual problems.
\begin{equation}
\tilde{\mathbf{x}}^{(d)} = \mathbf{P}^{(d)} \mathbf{x}^{(d)}
\end{equation}
where the transposed matrix
\begin{equation}
P^{(d)}_{ab} = \sum_{\alpha=0}^{M-1} \delta(a + \alpha, b) w(\alpha)
\end{equation}
and for brevity $w(\alpha) \equiv w_\alpha(d)$.

We cannot simply solve for the original timestream as we have less blocks in the pseudo-timestream than the original one, that is $\mathbf{P}$ is a rectangular matrix with fewer rows than columns, and the problem in under-determined. In this case we can apply the Moore-Penrose pseudo-inverse to find the least-squares (maximum-likelihood) solution, with *minimum* power. In our case this is actually equivalent to solving using the right-inverse
\begin{equation}
\hat{\mathbf{x}} = \mathbf{P}^T (\mathbf{P} \mathbf{P}^T)^{-1} \tilde{\mathbf{x}} \; .
\end{equation}
For efficiency and accuracy we will break this up into two calculations: first solve for an intermediate vector $\mathbf{y}$
\begin{equation}
(\mathbf{P} \mathbf{P}^T) \mathbf{y} = \tilde{\mathbf{x}}
\end{equation}
and then project this to estimate the timestream
\begin{equation}
\hat{\mathbf{x}} = \mathbf{P}^T \mathbf{y}
\end{equation}

First, let's look at the first step, in particular calculating what the matrix $\mathbf{P} \mathbf{P}^T$ is
\begin{align}
(\mathbf{P} \mathbf{P}^T)_{ab} & = \sum_{c} P^{(d)}_{ac} P^{(d)}_{bc} \\
& = \sum_b \sum_{\alpha, \beta} \delta(a + \alpha, c) \delta(b + \beta, c) w(\alpha) w(\beta) \\
& = \sum_{\alpha, \beta} \delta(a, b + \beta - \alpha) w(\alpha) w(\beta)
\end{align}
As this matrix is symmetric by construction, let's consider only the lower triangle, which corresponds to $b \ge a$, which, combined with the delta-function implies that $\beta > \alpha$. Given that $\beta - \alpha$ has a limited range, the matrix is band diagonal. To make life easier let's re-index this summation, defining $k = \beta - \alpha$ which runs from $0 \ldots M - 1$, and indexes which diagonal we are on, and $l = \alpha$, which runs from $0 \ldots M - 1 - k$. Taking this into account we can see that,
\begin{equation}
(\mathbf{P} \mathbf{P}^T)_{ab} = \sum_{k=0}^{M - 1} \delta(a, b + k) \left[\sum_{l=0}^{M - 1 - k} w(l) w(k+l) \right]
\end{equation}

```python
# Routine wrapping Lapack dgbmv
def band_mv(A, kl, ku, n, m, x, trans=False):
    import dgbmv
    
    y = np.zeros(n if trans else m, dtype=np.float64)
    
    lda = kl + ku + 1
    
    dgbmv.dgbmv('T' if trans else 'N', m, kl, ku, 1.0, A, x, 1, 0.0, y, 1)
    
    return y
```

```python
def inverse_pfb(ts_pfb, ntap, window=sinc_hanning):
    """Invert the CHIME PFB timestream.
    
    Parameters
    ----------
    ts_pfb : np.ndarray[nsamp, nfreq]
        The PFB timestream.
    ntap : integer
        The number of number of blocks combined into the final timestream.
    window : function (ntap, lblock) -> np.ndarray[lblock * ntap]
        The window function to apply to each block.
    """
    
    # Inverse fourier transform to get the pseudo-timestream
    pseudo_ts = np.fft.irfft(ts_pfb, axis=-1)
    
    # Transpose timestream
    pseudo_ts = pseudo_ts.T.copy()
    
    # Pull out the number of blocks and their length
    lblock, nblock = pseudo_ts.shape
    ntsblock = nblock + ntap - 1
    
    # Coefficients for the P matrix
    coeff_P = window(ntap, lblock).reshape(ntap, lblock)  # Create the window array

    # Coefficients for the PP^T matrix
    coeff_PPT = np.array([ (  coeff_P[:, np.newaxis, :]
                            * coeff_P[np.newaxis, :, :] ).diagonal(offset=k).sum(axis=-1)
                           for k in range(ntap) ])
    
    rec_ts = np.zeros((lblock, ntsblock), dtype=np.float64)
    
    for i_off in range(lblock):

        # Create band matrix representation of P
        band_P = np.zeros((ntap, ntsblock), dtype=np.float64)
        band_P[:] = coeff_P[::-1, i_off, np.newaxis]

        # Create band matrix representation of PP^T (symmetric)
        band_PPT = np.zeros((ntap, nblock), dtype=np.float64)
        band_PPT[:] = coeff_PPT[::-1, i_off, np.newaxis]

        # Solve for intermediate vector
        yh = la.solveh_banded(band_PPT, pseudo_ts[i_off])

        # Project into timestream estimate
        rec_ts[i_off] = band_mv(band_P, 0, 3, ntsblock, nblock, yh, trans=True)
        
    # Transpose timestream back
    rec_ts = rec_ts.T.copy()
    
    return rec_ts
```

## Test the PFB Inverse

```python
ntime = 2**20

# Generate a white noise timestream
ts = np.random.standard_normal(ntime)

# Perform the PFB
spec_pfb = pfb(ts, 65)

# Perform the inverse
rts = inverse_pfb(spec_pfb, 4)
```

## Missing Frequencies

One important effect that we are interested in is how well do we recover the timestream if we are missing frequency channels. We might worry that the effect of this frequency filtering of the PFB ends up being non-local in the frequency spectrum of the recovered timestream. An easy way of investigating this is to look at the cross correlation of the recovered signal with the input signal. In particular, we can use an FFT to channelise both timestreams, and look at the cross correlation as a function of channel (where we can normalise by the auto-correlation of the input).

One thing to note, to see the effective bandpass at higher resolution, we will use longer transforms to increase the frequency resolution of the FFT.

```python
def xcorr_channels(ts1, ts2, lblock=1024, skip=5):
    
    w = np.hanning(lblock)

    # Skip a number of blocks at the beginning and end, as we know that these are very noisy
    skip = 10

    # FFT the timestreams to get freq channels
    ch1 = np.fft.rfft(ts1.reshape(-1, lblock) * w, axis=1)
    ch2 = np.fft.rfft(ts2.reshape(-1, lblock) * w, axis=1)

    # Perform the various correlations
    c_11 = (ch1 * ch1.conj())[skip:-skip].mean(axis=0)
    c_12 = (ch1 * ch2.conj())[skip:-skip].mean(axis=0)
    c_22 = (ch2 * ch2.conj())[skip:-skip].mean(axis=0)
    
    return c_11, c_22, c_12
```

```python
ntime = 2**22

# Generate a white noise timestream
ts = np.random.standard_normal(ntime)

# Perform the PFB
spec_pfb = pfb(ts, 65)

# Zero a subset of channels
tchan = 30
spec_pfb[:, tchan:] = 0.0

# Perform the inverse
rts = inverse_pfb(spec_pfb, 4)

c_ii, c_oo, c_oi = xcorr_channels(ts, rts, lblock=1024, skip=10)
```

```python
plot(c_oi / c_ii, label='Cross-corr')
plot(c_oo / c_ii, label='Auto-corr', ls=':')

axvline(8*(tchan-0.5), color='k', ls='--')
legend()

xlabel('Hi-res channel')

ylim(0, 1.5)
#axvline(8*30)
```

As hoped, we find the cross correlation is one for frequency channels below the cutoff, and zero above, giving a bandpass which is what we expect.

However, if we look at the normalised auto-correlation we can see that there are large amounts of uncorrelated power appearing at all frequencies. We can interpret this as added reconstruction noise. This is dependent on what the window function used during the PFB was. For instance if we switch from the sinc-hanning window to a pure hanning window we can see that the noise is substantially decreased.

```python
ntime = 2**22

# Generate a white noise timestream
ts = np.random.standard_normal(ntime)

# Define new window function
wf = lambda nt, lb: np.hanning(nt * lb)

# Perform the PFB
spec_pfb = pfb(ts, 65, window=wf)

# Zero a subset of channels
tchan = 30
spec_pfb[:, tchan:] = 0.0

# Perform the inverse
rts = inverse_pfb(spec_pfb, 4, window=wf)

c_ii, c_oo, c_oi = xcorr_channels(ts, rts, lblock=1024, skip=10)
```

```python
plot(c_oi / c_ii, label='Cross-corr')
plot(c_oo / c_ii, label='Auto-corr', ls=':')

axvline(8*(tchan-0.5), color='k', ls='--')
legend()

xlabel('Hi-res channel')

ylim(0, 1.5)
#axvline(8*30)
```

## Quantisation

Another big question is how does quantisation of the PFB output affect our ability to reconstruct the timestream.

First let's write a routine to perform the quantisation.

```python
def quantise(arr, levels, ref):
    """Quantise a timestream.
    
    Parameters
    ----------
    arr : np.ndarray
        Array to quantise.
    levels : integer
        Number of levels in output timestream.
    ref : scalar
        The size of the level spacing (i.e. the value for the first level).
        
    Returns
    -------
    ts : np.ndarray
        Quantised array.
    """
    
    return np.maximum(np.minimum(np.rint(arr / ref), levels), -levels) * ref
```

Initially we will just look at the effects of quantisation on a timestream before the PFB.

```python
ntime = 2**22

# Generate a white noise timestream
ts = np.random.standard_normal(ntime)

qts = quantise(ts, levels=8, ref=0.5)

c_ii, c_oo, c_oi = xcorr_channels(ts, qts, lblock=1024, skip=10)

plot(c_oi / c_ii, label='Cross-corr')
plot(c_oo / c_ii, label='Auto-corr', ls=':')

legend()

xlabel('Hi-res channel')

ylim(0, 1.5)
#axvline(8*30)
```

As we can see, the cross correlation at all frequencies is essentially perfect, but a small amount of noise, giving an increased extra variance has appeared. Though we won't look in more detail here, we should note that this is dependent on the rounding mode, and shifting from rounding to the nearest integer, to truncation towards zero actually reduced the auto power, rather than adding it.

Let's now try performing a truncation after the PFB, and then performing the reconstruction.

```python
ntime = 2**22

# Generate a white noise timestream
ts = np.random.standard_normal(ntime)

# Define new window function
wf = lambda nt, lb: np.hanning(nt * lb)

# Perform the PFB
spec_pfb = pfb(ts, 65)

# Quantise data
spec_q = quantise(spec_pfb, 8, 4.0)

# Perform the inverse
rts = inverse_pfb(spec_q, 4)

c_ii, c_oo, c_oi = xcorr_channels(ts, rts, lblock=1024, skip=10)

plot(c_oi / c_ii, label='Cross-corr')
plot(c_oo / c_ii, label='Auto-corr', ls=':')

axvline(8*(tchan-0.5), color='k', ls='--')
legend()

xlabel('Hi-res channel')

ylim(0, 3.5)
#axvline(8*30)
```
